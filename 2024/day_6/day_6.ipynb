```python
import sys
import copy
```

```python
sys.path.append("../")
```

```python
file_path = 'input_ag.txt'
```

```python
# Read file content
raw_matrix = []
with open(file_path, 'r') as file:
    for line in file:
        raw_matrix.append(line.strip('\n'))
```

```python
def turn_right(current_direction):
    if current_direction == (0,1):
        return (1,0)
    elif current_direction == (1,0):
        return (0,-1)
    elif current_direction == (0,-1):
        return (-1,0)
    elif current_direction == (-1,0):
        return (0,1)
```

```python
def move_forward(current_direction, current_position):
    if current_direction == (0,1):
        return (current_position[0]-1, current_position[1])
    elif current_direction == (1,0):
        return (current_position[0], current_position[1]+1)
    elif current_direction == (0,-1):
        return (current_position[0]+1, current_position[1])
    elif current_direction == (-1,0):
        return (current_position[0], current_position[1]-1)
```

```python
def get_starting_state(puzzle_matrix):
    rows = len(puzzle_matrix)
    cols = len(puzzle_matrix[0])
    
    for row in range(rows):
        for col in range(cols):
            if puzzle_matrix[row][col] == '^':
                return (row, col,(0,1))
            elif puzzle_matrix[row][col] == '>':
                return (row, col,(1,0))
            elif puzzle_matrix[row][col] == 'v':
                return (row, col,(0,-1))
            elif puzzle_matrix[row][col] == '<':
                return (row, col,(-1,0))
    return None
```

```python
rows = len(raw_matrix)
cols = len(raw_matrix[0])

starting_row, starting_col, starting_direction = get_starting_state(raw_matrix)
puzzle_matrix = copy.deepcopy(raw_matrix)
puzzle_matrix = [list(row) for row in puzzle_matrix]
```

```python
def draw_travelled_path(starting_row, starting_col, starting_direction, puzzle_matrix):
    visited_states = set()
    filled_puzzle_matrix = copy.deepcopy(puzzle_matrix)
    current_row, current_col = starting_row, starting_col
    current_direction = starting_direction
    distance_travelled = 0

    while 0 <= current_row < rows and 0 <= current_col < cols:
        state = (current_row, current_col, current_direction)
        if state in visited_states:
            return 1
        visited_states.add(state)
        filled_puzzle_matrix[current_row][current_col] = 'X'
        new_row, new_col = move_forward(current_direction, (current_row, current_col))
        if 0 <= new_row < rows and 0 <= new_col < cols:
            if filled_puzzle_matrix[new_row][new_col] == '#':
                current_direction = turn_right(current_direction)
            else:
                current_row, current_col = new_row, new_col
                distance_travelled += 1
        else:
            current_row, current_col = new_row, new_col
    return filled_puzzle_matrix, distance_travelled, current_row, current_col, current_direction
```

```python
filled_matrix, distance_travelled, current_row, current_col, current_direction = draw_travelled_path(starting_row, starting_col, starting_direction, puzzle_matrix)
```

```python
for row in filled_matrix:
    print("".join(row))
```

```python
flattened_string = "".join("".join(row) for row in filled_matrix)

location_count = flattened_string.count('X')

print("Number of locations visited:", location_count)
```

```python
def test_obstacle_position(row, col, current_row, current_col,current_direction, reference_matrix):
    puzzle_matrix_test = copy.deepcopy(reference_matrix)
    puzzle_matrix_test[row][col] = '#'
    res = draw_travelled_path(current_row, current_col, current_direction, puzzle_matrix_test)
    if res == 1:
        print("Found loop at", row, col)
        return 1
    else:
        return 0
```

```python
def draw_travelled_path_with_obstacles(starting_row, starting_col, starting_direction, puzzle_matrix):
    
    reference_matrix = copy.deepcopy(puzzle_matrix)
    reference_matrix[starting_row][starting_col] = '.'
    
    rows, cols = len(reference_matrix), len(reference_matrix[0])
    count = 0
    for row in range(rows):
        for col in range(cols):
            if (row, col) != (starting_row, starting_col) and reference_matrix[row][col] != '#':
                if test_obstacle_position(row, col, starting_row, starting_col, starting_direction, reference_matrix):
                    count += 1
    print("Number of positions where adding an obstruction causes a loop:", count)
```

```python
draw_travelled_path_with_obstacles(starting_row, starting_col, starting_direction, puzzle_matrix)
```
